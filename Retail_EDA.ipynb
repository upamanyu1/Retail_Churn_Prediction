# EDA for customers_survey

```python
sns.countplot(customers_survey['Cust_Sex'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Customers of Grocery store by different sex')
```

```python
sns.countplot(customers_survey['Cust_Race'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Customers of Grocery store by different Race')
```

```python
sns.countplot(customers_survey['Cust_Rel_Status'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Customers of Grocery store by Status of Marriage')
```

```python
sns.countplot(customers_survey['Cust_Children'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Customers of Grocery store by number of children')
```

# EDA for customers dataframe

```python
plt.figure(figsize = (25, 8))
plt.style.use('fivethirtyeight')
ax = sns.countplot('Customer_City', data = customers, palette = 'dark')
ax.set_xlabel(xlabel = 'Cities', fontsize = 16)
ax.set_ylabel(ylabel = 'Count of Customers', fontsize = 16)
ax.set_title(label = 'Count of Customers in different cities', fontsize = 20)
plt.xticks(rotation=90)
plt.show()
```

# EDA for items dataframe

```python
plt.figure(figsize = (40, 8))
plt.style.use('fivethirtyeight')
ax = sns.countplot('Item_Name', data = items, palette = 'dark')
ax.set_xlabel(xlabel = 'Item_Name', fontsize = 16)
ax.set_ylabel(ylabel = 'Count of Items', fontsize = 16)
ax.set_title(label = 'Count of Items for each Item Name', fontsize = 20)
plt.xticks(rotation=90)
plt.show()
```

# EDA for Stores

```python
sns.countplot(stores['Store_City'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.xticks(rotation=90)
plt.title('No of stores in different cities')
```

```python
sns.countplot(stores['Store_Size'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.xticks(rotation=90)
plt.title('No of stores in with different sizes')
```

```python
# let's see how data is distributed for every column in CustomerData
plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in CustomerData:
    if plotnumber<=9 :     # as there are 9 columns in the data
        ax = plt.subplot(3,3,plotnumber)
        sns.distplot(CustomerData[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.show()
```
